## Final Code

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

titles =[]
prices =[]
final_urls=[]

## for no. of page do as follows:
for i in range(1,11): 
    
        ## Getting the responce for url
        base_url = 'https://speedcubeshop.com'
        url = 'https://speedcubeshop.com/collections/3x3-speed-cubes?page='+str(i)
        r = requests.get(url)  
        
        ## Parsing the data using beautifulsoup
        soup = bs(r.content,'html.parser')  
        box = soup.find('div',class_='product-collection products-grid row')
        
        ## Using the try and except error handeling to avoid the ('All parameters must be same length')...
        ##...error while making the dataframe.
        
        ## Getting the title_tags for products and if none is reurned 'No titles' will be returned.
        try:
            a_tags = box.find_all('a',class_='product-title change-text')
            len(a_tags)

            for i in range(0,len(a_tags),2):
                titles.append(a_tags[i].span.text.strip())
        except:
            titles = 'No titles'
        titles
        
        ## Getting the prices for products and if none is reurned 'No prices' will be returned.
        try:
            div_tags = box.find_all('div',class_='price-regular')
            len(div_tags)


            for i in range(0,len(div_tags),2):
                prices.append(div_tags[i].span.text)
        except:
            prices='No price'

        prices
        
        ## Getting the Links for products and if none is reurned 'No links' will be returned.
        try:
            all_product_links= box.find_all('a',class_='product-title change-text')
            all_product_links

            for i in range(0,len(all_product_links),2):
                       print(all_product_links[i])
            link0 = box.find('a',class_='product-title change-text')['href']
            print(base_url+link0)

            links=[]
            base_url='https://speedcubeshop.com'
            for url in all_product_links:
                links.append(base_url + url['href'])

            for i in range(0,24,2):
                final_urls.append(links[i])
        except:
            links = 'No links'
```

```python
## Creating a dictionary for extracted data to later convert it into a dafaframe.

data = {'Product_titles': titles,
        'Product_prices': prices,
        'Product_url': final_urls}
## Creating a dataframe from abov dictionary.

df = pd.DataFrame(data)

df
```
